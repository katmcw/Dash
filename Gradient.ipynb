```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/BANDSGO.csv")
df.loc[:, "time"] = pd.to_datetime(df.time)
df = df.sort_values(by="time")
```

```python
df_train = df.drop("time", axis=1)

df_class = df_train.copy()
df_class.loc[:, "salinity"] = 0
cutoff = df.psal.quantile(.5)
df_class.loc[df_class.psal > cutoff, "salinity"] = 1
df_class = df_class.drop("psal", axis=1)
```

```python
from sklearn.ensemble import GradientBoostingRegressor, GradientBoostingClassifier
from sklearn.model_selection import train_test_split
```

```python
inputs = [c for c in df.columns if "B" in c]

X = df_class.loc[:, inputs].values
y = df_class.loc[:, "salinity"].values

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2)
```

```python
#gb = GradientBoostingRegressor(random_state=4, n_estimators=5)#, n_estimators=100)
gb = GradientBoostingClassifier(random_state=4, n_estimators=100)
gb.fit(X_train, y_train)
```

```python
gb.score(X_train, y_train)
```

```python
gb.score(X_test, y_test)
```
